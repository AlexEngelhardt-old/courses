# Lack of Robustness of Expected Return Estimates

- Sample based information unfortunately is close to useless when it comes to expected return estimation

![Derp](problem.png)

- The black line: The entire 50% return could be just random fluctuation!
- The way out: Use **Bayesian statistics**
  - **Crucial**: We need to come up with *good* priors
  
### Shrinkage revisited

A simple prior:

The sample mean estimate $\hat{\mu}_i$ might be improved by shrinking the individual menas $\bar{\mu}_i$ towards the grand sample mean $\bar{\mu}$.

You want to estimate the mean return $\hat{\mu}_i$ for each stock in a universe of $N$ stocks.

Start with the noisy means:

$$\bar{\mu}_i = \frac{1}{T} \sum_{t=0}^{T-1} R^i_{t, t+1}$$

Compute the grand mean:

$$\bar{\mu} = \frac{1}{N} \sum_{i=1}^N \bar{\mu}_i$$

For the final estimate, shrink each noisy mean towards the grand mean by $\delta$:

$$\hat{\mu}_i = \delta \bar{\mu} + (1-\delta) \bar{\mu}_i$$

# Agnostic Priors on Expected Return Estimates

- Get an economically meaningful prior
- Anyone attempting to maximize his Sharpe Ratio needs some estimate of expected returns:

$$SR_p = \frac{\mu_p - r}{\sigma_p}$$

- **First agnostic prior**: Assuming all expected returns are equal
  - In this case, The maximum Sharpe ratio (MSR) portfolio is the global minimum variance (GMV) portfolio, because
  - Maximizing the sharpe ratio is equivalent to minimizing its denominator
- **Second agnostic prior**: Sharpe Ratios are all equal
  - i.e., The excess expected return is proportional to the asset's volatility:
  - $\mu_i - r_f = \lambda \sigma_i$
  - In this case, the MSR portfolio is "Max Div" (maximum Diversification Ratio):
  
$$SR_p = \frac{\sum_{i=1}^N w_i (\mu_i - r_f)}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}} = \lambda \frac{\sum_{i=1}^N w_i \sigma_i}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}}$$

# Using Factor Models to Estimate Expected Returns

- Factor models can provide us with meaningful economically motivated prior for expected returns

### CAPM-based expected return estimates

According to CAPM, differences in betas, and differences in betas alone, explain differences in expected returns.

- If the CAPM is the "correct" asset pricing model, then excess expected return is proportional to a stock's beta

$$\mu_i - r_f = \beta_i (\mu_M - r_f)$$

This is equivalent to imposing the prior that Treynor ratios are constant across assets

$$TR_i = \frac{\mu_i - r_f}{\beta_i} = \mu_M - r_f$$

$\mu_M - r_f$ is the *market risk premium*.

If the CAPM *were* the true model, then the implications for the Sharpe ratio would be:

$$SR_p = \frac{\sum_{i=1}^N w_i (\mu_i - r_f)}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}} = \frac{\sum_{i=1}^N w_i \beta_i (\mu_M - r_f)}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}} = (\mu_M - r_f) \frac{\sum_{i=1}^N w_i \beta_i}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}}$$

$\mu_i-r_f$ is $\beta_i(\mu_M-r_f)$. And maximizing the Sharpe ratio will be equivalent to maximizing $\sum^N_{i=1} w_i\beta_i$

And the implications for the derivation of the optimal portfoio would be:

$$\Rightarrow MSR = \max_{w_i} \frac{\sum_{i=1}^N w_i \beta_i}{\sqrt{\sum_{i,j=1}^Nw_iw_j\sigma_{ij}}}$$

- This is relatively simple, but unfortunately we don't live in a one-factor world.
- **The CAPM is likely not the "true" asset pricing model**

# Extracting Implied Expected Returns

- We're starting with the Black-Litterman (BL) model here already
- The incorporation of **active views** in a meaningful way
- The first step in the BL procedure: Extract implied expected return estimates
- The question at the very focus of the BL model:
  - Assuming you have available a set of active views (opinions of active managers about expected returns), how can you leverage them for your portfolio construction?


- BL model uses the Market portfolio, which is the true optimal portfolio according to the CAPM, as an *anchor point*, a benchmark, a starting point in your portfolio construction.
- BL is a Bayesian analysis application to portfolio construction
- It leads to intuitive portfolios with sensible weights

----

### Neutral / implied expected returns

- The neutral prior distribution is obtained by reverse engineering, assuming market or benchmark is the optimal portfolio

**Engineering:**

- Your input: $(\mu_i, \sigma_{ij}), i,j=1,\ldots ,N$
- Obtain output, by portfolio optimization (e.g. Max-Sharpe Ratio): $(w_i^*), i=1,\ldots ,N$, optimal portfolio weights

**Reverse Engineering:**

- We'll take as input not $\mu,\sigma$, but our input will be: $(\sigma_{ij}, w_i^\text{benchmark}), i,j=1,\ldots,N$
- By reverse portfolio optimization, you'll find out the values of the implied expected returns $\Pi = (\mu_i^\text{implied})$ that would support your benchmark as an *optimal* portfolio.
- Those implied returns will form the *prior* of the BL model


Assume two uncorrelated stocks with volatility 10% and 15% respectively. Further assume that the risk-free rate is 0%. What can we say about the neutral expected returns consistent with an equally-weighted benchmark portfolio?

The second stock must have an expected return 2.25 higher than the expected return of the first stock

We know that for uncorrelated stocks the Sharpe ratio maximizing weights are proportional to $(\mu_i-r)/\sigma_i^2$. In this case and since the risk-free rate $r=0\%$, we obtain: $\mu_1\sigma_1^2 = 50\% = \mu_2 / \sigma_2^2$. As a result, we have that $\mu_2/\mu_1 = \sigma_2^2 / \sigma_1^2 = 0.15^2 / 0.10^2 = 2.25$. In other words, the second stock must have an expected return 2.25 higher than the expected return of the first stock so as to justify optimally receiving the same allocation despite having a higher volatility.

----

$\Pi$, the vector of expected returns, is an anchor point, around which we add some uncertainty $\epsilon$. 

The combination of priors (i.e. these implied/expected returns) and manager's active views will create a posterior which we'll use for portfolio construction

# Introducing Active Views

- The second component in the BL model. The manager's active views about the expected returns.
- Formally, an active view about the expected return of all assets within a portfolio $P$ has a *normal distribution* with mean vector $Q$ 


$$P\mu = Q + \epsilon^\nu$$

where $\epsilon^\nu \sim N(0,\Omega)$

Managers' confidence level in their *active views* is reflected in $\Omega$.

- $n$ is the number of assets
- $k \leq n$ is the number of views
- $P$ is a kxn matrix that identifies the assets involved in the views
- $Q$ is a k-vector of the expected returns on these portfolios
- $\Omega$ is a kxn matrix of error terms in the views (confidence levels)

The BL model is summarized in **one (ugly) equation**:

$$\bar{\mu} = \left[ (\tau \Sigma)^{-1} + P'\Omega^{-1} P \right]^{-1} \left[ (\tau \Sigma)^{-1} + P'\Omega^{-1} Q \right]$$

It says: I'll combine expected implied returns $\Pi$ (which come with some uncertainty $\epsilon^e$):

$\mu = \Pi + \epsilon^e$

where $\epsilon^e \sim N(0,\tau \Sigma)$

I'll combine them with the *active views* $Q$ (which also come with uncertainties $\epsilon^\nu$):

$$P\mu = Q + \epsilon^\nu$$

where $\epsilon^\nu \sim N(0,\Omega)$

And in a Bayesian frame work the new expected returns (the *posterior* return vector) is the BL-formula, a complex mixture of the benchmark implied expected return, and the active views, and the confidence you have in those two values.

(If the manager has zero confidence in his views, your portfolio will converge towards the benchmark)

----

Assume that a manager’s active views for stock 1 return is 15% and it is 10% for stock 2 return.

This doesn't imply that stock 1 should be overweighted with respect to the benchmark portfolio weights. Regarding the first statement: it can happen that the benchmark implied expected return for stock 1 are is higher than 15%, in which case stock 1 should be underweighted compared to the benchmark, and it can also happen that the benchmark implied expected return for stock 2 is lower than 10% in which case stock 2 should be overweighted compared to the benchmark. Regarding the second statement: in the absence of information about volatilities of these stocks and their correlations with the rest of the portfolio constituents, a higher return is not a sufficient condition for a higher allocation.

# Black-Litterman Analysis (1999)

- BL combines implied returns and active views to generate **BL expected returns**
- Step 1: Extract market implied expected return
  - By assuming the cap-weighted portfolio is the max Sharpe ratio portfolio, and we use those weights as inputs, then derive what should be the implied values for the expected returns that justify the cap-weighted index as the MSR portfolio.
  - These implied returns are very highly correlated (99%) with the CAPM-based expected return estimates
  - (you don't have to use the cap-weighted index, you can use equally-weighted too)
- Step 2: Introduce active views
  - Either in absolute terms ("tech will return 20%") or relative terms ("Food will outperform books by 1%")
  - The confidence levels for the views are proportional to the variance of the prior
  
  
- BL is more typically used in the context of asset allocation decisions rather than stock picking decisions.
- It's a nice tool for investors that wish to incorporate active views into their investment decisions.

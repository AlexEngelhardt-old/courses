In recent years there has been increasing recognition of the importance of factor models in portfolio construction and analysis.

This first module provides an in-depth introduction to the most popular factor models used in investment practice, starting with William Sharpe’s **Capital Asset Pricing Model (CAPM)**, and then moving on multi-factor models.

We also discuss **style analysis**, which can be regarded as a constrained form of factor model, and which has been applied to the analysis of the performance of active managers.

Finally, we discuss the shortcomings of cap-weighted equity indices, including their inefficient exposure to rewarded risk factors, and we also discuss the recent “**smart beta**” approach, which has been introduced as a formal methodology allowing investors and asset managers to build more efficient portfolios. 

# Introduction to factor investing

- The traditional choice was between **active and passive**.
  - Passive: Don't try and outperform the index
- New: Factor investing lies in-between. Build a systematic set of rules to try and outperform the S&P. 
  - Call it "Low cost active". It's not just the S&P500, *but* it's passive in the sense that there are no human, subjective decisions in there (after constructing the rules)
- Indicization refers to the trend of creating new indices that capture the portion of active management that is rules based and systematic, and in the long run should outperform the cap-weighted benchmark.

A factor investor looks at active funds and tries to figure out what are the **drivers** that provide that excess return (e.g. small size, or political risk). Then he figures out how to get them for himself, inexpensively.

If you're willing to expose yourself to this factor, you'll be rewarded with a risk premium

# Factor models and the CAPM

**Factor model**: A decomposition of the return $r$ pf an asset into a sum of factor premia:

$$R_i = \beta_1 f_1 + \beta_2 f_2 + \ldots + \alpha + \epsilon $$

- $\alpha$: Some fixed term (this later becomes the **added value of the active manager**)
- $\epsilon$: Some error term

### CAPM: The best known factor model

- The CAPM includes much more than just a factor model, but it embeds one
- It says: The excess return of a stock over the risk free rate is: The excess return of *the market* over the risk free rate *times* the beta of the stock.
  - A stock's beta is its covariance to the market divided by the market's variance
  
$$ E(r_i)-r_f = \beta_i (E(r_m) - r_f)$$

$$\beta_i = \frac{\text{Cov}(r_i, r_m)}{\text{Var}(r_m)} $$

- A beta of zero means this stock does not carewhat the market does. All variation will be in the error term.
- In the CAPM, the $\alpha$ from above is zero.

### Behavioral anomalies

- Low-volatility stocks tend to outperform high-vol stocks. This is **exactly the opposite of what the CAPM says**.
- Momentum is a behavioral factor too
- Value vs. growth is rationally explainable, but has a behavioral component too (see the "We like stories" chapter in Montier's "Little Book of Behavioral Investing")

# Multi-factor models and the Fama-French model

- Fama-French is the best-known extension to the CAPM
  - It explains e.g. why value stocks outperform growth stocks
- It adds two factors to the CAPM:
  - **Value**: Book to price ratio (a measure of **value**: high ratio means *value stock*)
  - **Size**: The lowest 10% re. market cap outperform the largest 10%
- Those two are not explainable by the CAPM's $\beta$

$$ E(r_i) = r_f + \beta_{i,\text{MKT}} E(r_m-r_f) + \beta_{i,\text{SMB}} E(\text{SMB}) + \beta_{i,\text{HML}} E(\text{HML})$$

### Carhart model (four factors)

- Adds **Momentum**: Previous 12 months' winners vs. losers.

### Other factors

- Low-vol beats high-vol (That's *probably* a factor too)
- High-quality beats low-quality

- SOTA opinion: low-vol, value, momentum, and quality are treated as factors
  - Size is not treated as a factor, but as a *universe choice*
  - Why is value not a factor (anymore)?

# Factor benchmarks and Style analysis

- You can use factors as a diagnostic: To understand and decompose, to learn how an asset manager is achieving his returns
- Consider a portfolio with a $\beta$ of 1.3:
    $$E(r_i-r_f) = \alpha + 1.3 \cdot E(r_m-r_f)$$
- You can reproduce this portfolio by borrowing 30ct and putting \\$1.30 in the normal (i.e. $\beta=1$) market:
    $$E(r_i) = \alpha + (-0.3 r_f + 1.3 E(r_m))$$

If you're getting anything in excess of that, **then** the $\alpha$ term becomes the value added by the manager. Otherwise, the manager is just harvesting factor premia.

Assume the risk free rate is 1% per year and the Stock Market returned 11% in a given year. An Active Manager “beat the market” and generated a 14% return and had a Beta of 1.3. Then the manager generated **zero** alpha.

### Sharpe's style analysis (1992)

- Decompose a manager's return into factors
- Allow the benchmark to change over time
- Factor/style loading are re-estimated every period using data up until the current period
- Sliding window of 1-3 years allows for style drift of the manager
- This can be considered a *custom benchmark* for the manager


$$r_m = w_1 r_{i1} + w_2 r_{i2} + w_3 r_{i3} + \alpha + \epsilon$$


- You can solve for $w_j$ and $\alpha$ (and $\epsilon$?) with *quadratic programming*, to find out a manager's $\alpha$
  - You can find out his factor exposures given only his set of returns!
  - i.e., you don't need his asset allocation.
  - Look at the returns and be able to say "ok he's a value manager"

# Shortcomings of cap-weighted indices

Summary: **Cap-weighted indices like the S&P500 are inefficient portfolios and bad benchmarks**

- The default approach to benchmark managers is to use a cap-weighted index
- They are simple and easy to understand, but inefficient benchmarks
  - They don't generate the kind of risk-adjusted performance you could expect from an **efficient** allocation to the underlying stocks.

> Cap-weighted stock portfolios are inefficient investments. [...] Even the most comprehensive cap-weighted portfolios occupy positions inside the efficient set [i.e. frontier].
>
> -- Haugen and Baker (1991)

- The S&P500 is inefficient (ex-post): It's way inside the efficient frontier that you obtain from *efficiently* combining its constituent stocks.
- CW indices are heavily concentrated, thus poorly diversified
  - Nokia was once over 50% of the Finnish stock market
- EW (equally weighted) dominates CW (capitalization-weighted) with regards to the Sharpe ratio
    - (Platen and Rendek 2010)
    - This is a naive approach, but already works well
- Even monkeys are better! Random weighting outperforms cap-weighted :)
- Other alternatives that outperform the cap-weighted portfolio on a risk-adjusted basis:
  - Equally weighted benchmarks
  - Minimum variance benchmarks
  - Risk parity benchmarks

# From cap-weighted benchmarks to smart-weighted benchmarks

- Two reasons why CW index is inefficient
  1. They're concentrated and inefficiently diversified; they contain an excess of unrewarded risk
    - use a smarter weighting scheme
  2. They provide inefficient exposure to rewarded systematic risks
    - **They're biased towards large-caps (because they're larger, i.e. higher market cap) and growth stocks (because they're more expensive, i.e. higher market cap)**
    
Fix these problems by using **smart factor benchmarks/indices**

Steps:

1. Select your desired factor exposures
    - Value
    - Small cap
    - Momentum
    - Low-volatility
2. Select your preferred weighting scheme(s)
    - Equally weighted
    - Efficient minimum variance
    - Risk Parity
    - etc...
